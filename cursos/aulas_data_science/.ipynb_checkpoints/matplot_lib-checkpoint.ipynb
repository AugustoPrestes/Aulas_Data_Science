# MatplotLib

## Introdução à biblioteca

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

```python
# Declarado as variaveis para o eixo X e eixo Y

eixo_x_dias = np.random.randint(31, size=7)
eixo_y_temp_max = np.random.uniform(23, 27, size=7)
eixo_y_temp_min = np.random.uniform(18, 22, size=7)
```

```python
# Ordenando as variavéis do menor para o maior
eixo_x_dias.sort()
```

```python
# Exemplos de estilos pré definidos
style_list = plt.style.available
print(style_list)
```

```python
# Visualizando o gráfico de exemplo em todos os estilos pré-definidos
for style in style_list:
    print(style)
    plt.style.use(style)
# Editando a visualização do gráfico
    plt.title("Mínima e Máxima da semana")
    plt.xlabel("Dia")
    plt.ylabel("Temperatura máxima")
    
    
# Criando e visualizando o gráfico
    # O parametro 'marker' representa o estilo de markador que utilizaremos
    plt.plot(eixo_x_dias, eixo_y_temp_max)
    plt.plot(eixo_x_dias, eixo_y_temp_min)
    
    # Inserindo as legendas do gráfico
    plt.legend(["Temp. Máx", "Temp. Min"])
    
    plt.show()
```

## Gráfico de Linha

```python
# Visualizando o gráfico de exemplo em todos os estilos pré-definidos
# Editando a visualização do gráfico
plt.style.use("classic")

plt.title("Mínima e Máxima da semana")
plt.xlabel("Dia")
plt.ylabel("Temperatura máxima")


# Criando e visualizando o gráfico
plt.plot(eixo_x_dias, eixo_y_temp_max, label="Temp. Máx")
plt.plot(eixo_x_dias, eixo_y_temp_min, label="Temp. Min")

# Plotando a legenda
plt.legend(loc="best")

# Ajustando os axis do gráfico
plt.axis([1, 31, 15, 30])

plt.show()
```

## Gráfico de Barras

```python
# Declarando as variaveis
cidade_a_temp = np.random.uniform(23, 27, size=7)
cidade_b_temp = np.random.uniform(23, 27, size=7)
cidade_c_temp = np.random.uniform(23, 27, size=7)

datas = np.random.randint(31, size=7)
```

```python
datas.sort()
```

```python
# Gráfico de Barras Sobrepostas
plt.bar(datas, cidade_a_temp, width=1, label='Cidade A')
plt.bar(datas, cidade_b_temp, width=1, label='Cidade B')
plt.bar(datas, cidade_c_temp, width=1, label='Cidade C')

plt.legend()

plt.show()
```

```python
# Gráfico de Barras Lado a Lado

# Gerando as posições
posi_a = list(range(len(datas)))
posi_b = [posi + 0.25 for posi in posi_a]
posi_c = [posi + 0.25 for posi in posi_b]

# Atribuindo os valores as posições das barras
plt.bar(posi_a, cidade_a_temp, width=0.25, label='Cidade A')
plt.bar(posi_b, cidade_b_temp, width=0.25, label='Cidade B')
plt.bar(posi_c, cidade_c_temp, width=0.25, label='Cidade C')

# Gerando a legenda
plt.legend()

# X label
plt.xticks(ticks=posi_a, labels=datas)

plt.show()
```

Vizualização importante para comparar os valores de variáveis categóricas

## Séries Temporais com Matplotlib

```python
from datetime import datetime, timedelta
from matplotlib import dates as plt_dates
```

```python
# Declarando as variáveis

# datas
dates = [
    datetime(2020, 1, 2),  
    datetime(2020, 1, 4),    
    datetime(2020, 1, 5),   
    datetime(2020, 1, 8),   
    datetime(2020, 1, 11),    
    datetime(2020, 1, 13),   
    datetime(2020, 1, 17),  
    datetime(2020, 1, 24),    
    datetime(2020, 1, 26),    
    datetime(2020, 1, 30)    
]

y = list(range(1, 11, 1))
```

```python
# Configurando a visualização
plt.style.use('seaborn-v0_8')

# Criando a visualização
plt.plot(dates, y, marker='o', linestyle='')


# Formatando a vizualização das datas no eixo X
plt.gcf().autofmt_xdate()

# Formatando as datas para o eixo X
date_format = plt_dates.DateFormatter('%d/%m/%Y') # Formato Brazil
# date_format = plt_dates.DateFormatter('%b, %d %Y') # Formato USA
plt.gca().xaxis.set_major_formatter(date_format)

plt.tight_layout()
plt.show()
```

## Adicionando os valores aos rotulos da visualização

```python
# Declarando variáveis
days = ['seg', 'ter', 'qua', 'qui', 'sex', 'sab', 'dom']

temp = [22, 25, 27, 24, 26, 30, 29]
```

### Exemplo: Gráfico de barras

```python
# Gerando o gráfico e armazenando em uma variável
bar_plot = plt.bar(days, temp)

# Passando a posição dos valores de cada barra
plt.bar_label(bar_plot, labels=temp, label_type="edge")

plt.show()
```

## Annotate
Maneira de adicionar comentarios à visualização, para agregar o Strorytelling

```python
# Dados de Exemplo
x = range(10, 51, 10)
y = range(1, 6)
print(list(zip(x, y)))

plt.style.use("seaborn-v0_8")

# Criando a visualização
plt.plot(x, y, marker='o', linestyle='--')

# Annotate
plt.annotate(
    'Segundo marcador', # Texto da Anotação
    xy=(x[1], y[1]), # Ponta da seta
    horizontalalignment='center', # left, center, right
    verticalalignment='bottom', # top, center, bottom  
    arrowprops={'facecolor': 'grey', 'shrink': 0.05}, # Propriedades da seta (shrink é o parametro de afastamento de ambas as pontos)
    xytext=(x[1]+0.5, y[1]-0.8) # Posição do texto
)

plt.show()
```
