## BoxPlot - Para visualizar outliers
* Gráfico de caixa, que contém alguns parametros estatisticos:
    1. IQR - Inter Quartil;
    2. Mediana;
    3. Outliers.

* Como calcular o L.I (Limite Inferior) e L.S (Limite Superior):
    1. **L.I.** = $Q_1 - 1.5 * IQR$
    2. **L.S.** = $Q_3 + 1.5 * IQR$
    3. Caso o **L.I** sejá menor que o meu valor minimo, utilizaremos o valor minimo como Limite, porém se o valor minimo for menor que o meu **L.I** consideramos esse número como um outlier. Caso o **L.S** sejá maior que o valor máximo, utilizaremos o valor máximo como limite, porém se o valor máximo for maior que o meu **L.S** consideramos esse número como um outlier.

# Probabilidade
Forma de mensurar a aleatoriedade 

## Eventos

Um evento é o resultado observado de um experimento. E um evento de interesse é o evento que esta sendo observado:

* Exemplo: <br> 
    * **Experimento:**  Lançar uma moeda e observar qual lado ficou para cima.<br>
    * **Evento Possível:**  Dar *Cara* ou *Coroa*.

---

## Espaço amostral
É o espaço onde todas as possibilidades de evento podem ocorrer.

* Exemplo: <br>
    * **Experimento:**  Lançar uma moeda e observar qual lado ficou para cima.<br>
    * **Evento Possível:**  Dar *Cara* ou *Coroa*.
    * **Espaço Amostral:**  {Cara, Coroa}.

---

## Probabilidade

Definimos a probabilidade de um evento acontecer como: $P(E)$

Para utilizar como exemplo temos alguns axiomas de probabilidade:

1.  $ 0 \leq P\left ( E_{i} \right ) \leq 1$ - A probabilidade sempre está entre 0 e 1.
2. $P(\Omega ) = 1$ - Ao menos **1** *evento* acontecera dentro do meu *espaço amostral*
3. Para todos eventos $E_{1}, E_{2}...$ mutuamente exclusivos temos: $P(\cup_{i = 1}^{\infty } E_{i}) = \sum_{i = 1}^{\infty}P(E_{i})$
    1. Exemplo com um dado de 6 lados:
        * $(E_1 = 1, E_2 = 2, E_3 = 3, E_4 = 4, E_5 = 5, E_6 = 6)\Rightarrow $
        * $P(E) = 1/6 $  
        * $P(\Omega) = \sum_{i = 1}^{6}P(E_{i})=1 $

### Como podemos calcular (inferir) a probabiliade?
1. Definir a probabilidade associada a cada evento;
2. Realizar experimentos;
3. Assumir uma função que gera os dados;
4. Experiências passadas.

## Probabilidade condicional

Probabilidade condicional é a probabilidade baseada em alguma informação previamente conhecida. Isto é, predizer a respeito do evento **A**, quando já sabemos a ocorrência do evendo **B**. Denotaremos por $P(A|B)$, leia-se probabilidade do evento **A** dado o evento **B**.

Dizemos ainda que, caso $P(A|B) = P(A)$, então podemos dizer que **A** e **B** são eventos independentes.  

Podemos dizer que a **Probabilidade Condicional** é como adição de informação, ou dica a respeito do evento de interesse.

* **Exemplo:** Probabilidade do resultado do lançamento de um dado ser 4, sabendo que o resultado foi par?

---

## Probabilidade Conjunta

A probabilidade conjunta é a probabilidade de dois ou mais eventos ocorrerem simultaneamente. Em um caso particular de apenas dois eventos, seguimos com a notação.

* $P(A\cap B) = P(A)P(B|A) = P(B)P(A|B)$

Repare que no caso de eventos independentes, temos:
* $P(A\cap B) = P(A)P(B) = P(B)P(A)$

**Exemplos:**
1. Lançar dois dados e aparecer o número 3 em ambos;
2. Lançar uma moeda 3 vezes e aparecer 3 caras;
3. Chover e terrelâmpagos.

--- 
## Teorema de Bayes
### $P(A\cap B) = P(B)P(A|B)$
### $\frac {P(A\cap B)}{P(B)} = P(A|B)$
### $\frac {P(A)P(B|A)}{P(B)} = P(A|B)$

## Distribuição Normal



### $f(x;\mu, \sigma ) = \frac{1}{\sqrt{2\pi \sigma }}e^{\frac{-1}{2}\left ( \frac{x-\mu }{\sigma} \right )^{2}}$

A média e a variância da amostra servem para ajustar os parâmetros de uma distribuição normal:

### Parametros em uma População
$\mu$ - É o parametro de **locação**<br>
$\sigma$ - É o parametro de **disperção**

### Parametros em uma Amostra
$\overline{x}$ - É o parametro de **locação**<br>
$s^{2}$ - É o parametro de **disperção**


Por isso a média é muito importante, pois com ela, podemos ajustar uma função de probabilidade.

Porém o que é utilizado são sempre os dados da amostra. Por isso é sempre importante entender que devemos gerar a amostragem aleatória mais de uma vez. E calcular o intervalo de confiança.

## Intervalo de Confiança

Nos traz informações sobre uma região de confiança onde acreditamos que o verdadeiro valor se encontra, com base em uma probabilidade associada. Retirada da média de amostras aleatórias.
